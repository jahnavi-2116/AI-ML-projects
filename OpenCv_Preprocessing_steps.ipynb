```python
import cv2

car = cv2.imread('/content/IMG_8157.jpeg')
```

```python
import matplotlib.pyplot as plt
```

```python
image_rgb = cv2.cvtColor(car, cv2.COLOR_BGR2RGB)
```

```python
# Display image
plt.imshow(image_rgb)
plt.title("car")
plt.axis('off')
plt.show()
```

```python
resized_image = cv2.resize(car, (224, 224))
plt.imshow(cv2.cvtColor(car, cv2.COLOR_BGR2RGB))
plt.title("Resized Image (224x224)")
plt.axis('off')
plt.show()
```

```python
gray_image = cv2.cvtColor(car, cv2.COLOR_BGR2GRAY)
plt.imshow(car, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```

```python
normalized = car / 255.0
plt.imshow(car)
plt.title("Normalized Image")
plt.axis('off')
plt.show()
```

```python
# Blurring and Denoising image
blurred = cv2.GaussianBlur(car, (5, 5), 0)
plt.imshow(cv2.cvtColor(car, cv2.COLOR_BGR2RGB))
plt.title("Gaussian Blurred Image")
plt.axis('off')
plt.show()
```

```python
# Edge detection
edges = cv2.Canny(car, 100, 200)
plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis('off')
plt.show()
```

```python
# thresholding image
ret, thresh = cv2.threshold(car, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresh, cmap='gray')
plt.title("Thresholded Image")
plt.axis('off')
plt.show()
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))

eroded = cv2.erode(thresh, kernel, iterations=1)
dilated = cv2.dilate(thresh, kernel, iterations=1)

plt.subplot(1,2,1), plt.imshow(eroded, cmap='gray'), plt.title('Eroded')
plt.subplot(1,2,2), plt.imshow(dilated, cmap='gray'), plt.title('Dilated')
plt.show()
```

```python
hsv_image = cv2.cvtColor(car, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_image)
plt.title("HSV Image")
plt.axis('off')
plt.show()
```
